```python
import pandas as pd
```

```python
#sample data
data=[10,12,12,13,12,11,10,250]
df=pd.DataFrame(data,columns=["value"])
```

```python
#step1 : Find q1 and q3
q1=df["value"].quantile(0.25)
q3=df["value"].quantile(0.75)
```

```python
#step 2: Calculate IQR
iqr=q3-q1
```

```python
#step 3: Define outlier boundaries
lower_bound=q1-1.5*iqr
upper_bound=q3+1.5*iqr
```

```python
print(lower_bound,upper_bound)
```

```python
#step 4: Identify Outliers
outliers=df[(df["value"]<lower_bound) | (df["value"]>upper_bound)]
```

```python
print("Data:\n", df)
print("\nQ1:", q1, " Q3:", q3, " IQR:", iqr)
print("\nOutliers:\n", outliers)
```

```python
import matplotlib.pyplot as plt

# Plot boxplot and histogram side by side
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# --- Boxplot ---
axes[0].boxplot(df["value"], vert=True)
axes[0].set_title("Boxplot of Values")
axes[0].set_ylabel("Value")

# --- Histogram ---
axes[1].hist(df["value"], bins=10, color="skyblue", edgecolor="black")
axes[1].set_title("Histogram of Values")
axes[1].set_xlabel("Value")
axes[1].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```
